```python
import numpy as np 
import pandas as pd 
from mpl_toolkits.mplot3d import axes3d, Axes3D
import matplotlib.pyplot as plt 
```

## adsorción

$q = C^{1/n} {K}$ <br>
$ln (q) = {\frac 1 n}ln (C)+ ln (K)$ <br>
$q = (\frac V M)(C_{0}-C)$

## filtración

$
	Q = K * \frac{{h_1}-{h_2}}{L} * A = K * \frac{\Delta h}{L} * A
$

## desinfección

<p>$  t^{2} = {\frac 2 k} Log \left( \frac {N_{0}} {N_{t}} \right) $  </p><br>

<p>$  t = \sqrt{{\frac 2 k} Log \left( \frac {N_{0}} {N_{t}} \right)} $  </p><br>

<p>$  t = \sqrt{{\frac 2 k} Log \left( {\frac {100}{100 - \% Mortalidad}} \right) }$  </p><br>

<p>$ { \% Mortalidad}  =    100 - \left( {\frac {100}{10^{kt^{2}/2}} }\right)  $  </p><br>

#ISOTERMAS

### caso 1:
1/n: 1.2 <br>
c: 0-1 <br>
k: 0-1

```python
C = np.linspace(0,10, 100)
slope = 1.2 #1/n
K = np.linspace(0,1,100)
```

```python
C1n = C**slope
```

```python
q = C1n*K
```

```python
C2, K2 = np.meshgrid(C, K)
```

```python
q2 = (C2**slope)*K2
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot3D(C1n, K, q, color="red")
ax.set_xlabel("$C^{1/n}$")
ax.set_ylabel("$K$")
ax.set_zlabel("$q$")
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")
ax.plot_surface(C2, K2, q2, rstride=1, cstride=1,cmap='copper', edgecolor='none',)
ax.set_xlabel("$C^{1/n}$")
ax.set_ylabel("$K$")
ax.set_zlabel("$q$")
ax.set_title("1/n = 1.2")
plt.show()
```

## Caso 3
C = 1 <br>
n: 0.5:1.5 <br>
k: 0:1

```python
C4 = 1
n4 = np.linspace(0.5,1.5,100)
k4 = np.linspace(0,1,100)
n4,k4 = np.meshgrid(n4,k4)
q = (C4**(1/n4)) * k4
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")
ax.plot_surface(1/n4, k4, q, rstride=1, cstride=1,cmap='twilight', edgecolor='none')
ax.set_xlabel("${1/n}$")
ax.set_ylabel("$K$")
ax.set_zlabel("$q$")
ax.set_title("C = 1")
plt.show()
```
