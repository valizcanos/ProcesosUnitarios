# **Ley de Darcy**



image.png

Fuente de la imagen:

https://www.youtube.com/watch?v=7awSAWXGLDg

<p>La ley de Darcy es un planteamiento matemtico que describe el flujo de un fluido en un medio poroso.</p>

<p>Cuando fluido entra en contacto con una superficie porosa, su presin hidrosttica va variando respecto a su desplazamiento ($L$). Especficamente la presin es mayor ($h_1$) en los puntos ms cercanos a la fuente (afluente) y menor ($h_2$) en los puntos ms cercanos al drenaje (efluente). A este cambio de presiones con respecto a $L$ se le conoce como gradiente hidrulico.</p>
<br>

image.png

Fuente de la imagen:

https://www.lifeder.com/ley-darcy/

\begin{equation}
	Q = K * i * A
\end{equation}

\begin{equation}
	Q = K * \frac{{h_1}-{h_2}}{L} * A = K * \frac{\Delta h}{L} * A
\end{equation}

<p>La relacin entre el tipo de medio poroso (arena, grava, arcilla, ...), el caudal ($Q$) y el gradiente hidrulico ($i$) permiten definir su conductividad hidrulica ($K$). $K$, representa la capacidad de cualquier tipo de medio poroso para permitir el paso del fluido a trves de l. </p><br>

<p>Al relacionar el $Q$ y el rea de seccin transversal ($A$) por la cual se desplaza el fluido se obtiene la velocidad. A esta velocidad se le conoce como velocidad de Darcy ($\vec V_{Darcy}$) o velocidad falsa, puesto que el fluido no circula por toda la seccin del medio.</p><br>

\begin{equation}
	\vec V_{Darcy} = \frac{Q}{A} = K * i 
\end{equation}

<p>La velocidad real ($\vec V_{real}$) se define a travs de su coeficiente de porosidad eficaz ($m_e$). La $m_e$ representa la cantidad de fluido que estara circulando por el por una fraccin del medio poroso. </p><br>

<p>Hay que tener el encuenta que la $\vec V_{Real}$ no es equivalente a la velocidad observada ($\vec V_{Observada}$). La $\vec V_{Observada}$, es aquella velocidad que se mide experimentalmente al  al cronometrar el tiempo de recorrido entre dos puntos. </p><br>

</p>En la prctica la $\vec V_{Observada}$ sera inferior a la calculada mediante ya que el agua ha tenido que recorrer una distancia mayor (ha recorrido $L_2$ y no $L_1$) y por lo tanto, el tiempo real de recorrido entre dos puntos puede ser ligeramente mayor al predicho mediante la expresin que representa a la $\vec V_{real}$. </p>

$\vec V_{Real} = \frac{\vec V_{Darcy}}{m_{e}}$

image.png

<p>La relacin entre la velocidad observada desde el exterior del medio poroso y la calculada a partir de Darcy ($\vec V_{Real}$) y de la porosidad eficaz ser as:</p> <br>

$\vec V_{Observada} = \frac{\vec V_{Real}}{Coef._{Tortuosidad}}$

<p>Ese coeficiente depende de la tortuosidad del medio poroso, y aproximadamente puede ser de 1.0 a 1.18 en arenas </p>

image.png

# **EJEMPLO 1:**
En un permemetro de laboratorio se quiere medir la conductividad hidrulica de una arena. La seccin del permemetro es de 340 $cm^2$ y la distancia entre las dos conexiones a los tubos piezomtricos ($\Delta l$) es de 45 cm. El caudal de salida es de 2.7 litros/minuto y la diferencia de altura entre las dos columnas de agua ($\Delta h$) es de 68 cm. Calcular la conductividad hidrulica en metros/da. 


\begin{equation}
	Q = K * \frac{{h_1}-{h_2}}{L} * A = K * \frac{\Delta h}{L} * A
\end{equation}

```python
Q1 = (2.7/1000)*(1/60) # Caudal en m^3 / segundo
Q1
```

```python
DeltaL1 = 45/100 # Longitud del medio poroso en metros, medida desde la posicin de los piezometros
DeltaL1
```

```python
Deltah1 = 68/100 # Cambio de altura medida en los piezometros (en metros)
Deltah1
```

```python
A1 = 340/10000 # rea de seccin de prememetro en m^2
A1
```

```python
K1 = (Q1*DeltaL1)/(A1*Deltah1) #En m/s
K1 = K1*86400 #m/d
K1
```

#**LEY DE DARCY PARA VARIAS CAPAS DE MEDIOS POROSOS**

En un medio estratificado, con frecuencia se produce el flujo a travs de varias capas, para ello se debe emplear un coeficiente global de conductividad hidrulica. Para este caso hay que tener en cuenta que el coeficiente global de conductividad va a cambiar si el flujo es horizontal ($K_h$) o perpendicular ($K_v$) al medio.

image.png

$K_{h}=\frac{\sum K_{i}b_{i}}{B}$

$K_{v}=\frac{B}{\sum \frac {b_{i}}{K_{i}}}$

$K_{i}:$ conductividad hidrulica de cada una de las capas<br>
$K_{h}:$ conductividad hidrulica horizontal equivalente<br>
$K_{v}:$  conductividad hidrulica vertical equivalente <br>
$b_{i}:$ espesor de cada una de las capas<br>
$B:$ = espesor total, suma de todos los espesores<br>

#**Estimacin de caudales a travs de diferentes capas porosas**

#Flujo horizontal

image.png

$Q= (\sum{-K_{i}*b_{i}*w})* \frac{\Delta h}{\Delta l} = (\sum{-K_{i}*b_{i}*w})*i$

## Caudal total horizontal que circula a travs de todas las capas <br>
$Q= (\sum{-K_{h}*B}*w)* \frac{\Delta h}{\Delta l} = (\sum{-K_{h}*B*w})*i$ <br>

w: representa el ancho en direccin a la cual circula el flujo

# Flujo vertical

image.png

$Q = -K_{v}*\frac{\Delta h}{B} * A$

$Q_{i} = -K_{i}*\frac{\Delta h_i}{b_{i}} * A$

$K_v = K_{i}*\frac{\Delta h_i}{b_{i}} * \frac{B}{\Delta h}$

$K_{v}=\frac{B}{\sum \frac {b_{i}}{K_{i}}}$

#**CONDUCTIVIDAD**
La conductividad $K$ depende del medio y de las caractersticas del propio fluido. El factor $K$ puede descomponerse as: 

$K = k \frac{}{}$

Donde, $k$ es la permeabilidad intrnseca o absoluta (depende slo del medio poroso) 

$:$ peso especfico del fluido <br>
$\mu:$ viscosidad dinmica del fluido <br>

$K = k \frac{g}{}$

$:$viscosidad cinemtica del fluido

La variacin de la temperatura afecta la conductividad de un fluido en un medio poroso, por ejemplo, consideremos la ecuacin anterior de un mismo medio a dos temperraturas diferentes:

$K_{T1} = k \frac{g}{_{T1}}$ <br>
$K_{T2} = k \frac{g}{_{T2}}$ <br>
$k{g} = K_{T1}*{_{T1}} = K_{T2}*{_{T2}}$ <br>
$K_{T1} = {_{T2}} \frac{K_{T2}}{_{T1}}$ <br>

## **La relacin entre el caudal y el gradiente hidrulico**

valor de K es muy bajo o cuando las velocidades del flujo son muy altas. si se aplicara la Ley de Darcy para calcular el flujo a travs de una formacin
arcillosa, el caudal que s obtendra sera muy bajo, pero en la realidad ser nulo, no habr circulacin de agua si no se aplican unos gradientes muy elevados. En el caso contrario, si el agua circula a gran velocidad, el caudal es directamente proporcional a la seccin y al gradiente, pero no linealmente proporcional, sino que la funcin sera potencial: 

$\vec V = -K \left( \frac{dh}{dL} \right)^{n}$

Donde el exponente n es distinto de 1. 

Para estudiar este lmite de validez de la ley de Darcy se aplica el nmero de Reynolds. Donde  el dimetro de la conduccin sera el dimetro medio
de los granos del medio poroso y la velocidad empleada sera la **velocidad de Darcy**.

$R_{e}= \frac{\rho \vec V d}{\mu} = \frac{\vec V d}{}$

d = dimetro medio de partcula (m)

$\mu$ = viscosidad dinmica ($Pascal\cdot m = \frac{Kg}{m^2 \cdot s}$ ) 

$$= viscosidad cinemtica ($\frac{\mu}{\rho} = \frac{m^2}{s}$)

$\rho$= densidad ($\frac{kg}{m^3}$)

$\vec V$ = $\vec V_{Darcy}$ = velocidad de Darcy ($\frac{m}{s}$)

$\vec V_{Darcy} = \frac{R_{e} \mu }{\rho d} = \frac{R_{e} \upsilon }{d} = K i$

En la realidad es imposible conocer el grado de turbulencia del flujo a travs de un medio poroso. Sin embargo, experimentalmente se ha observado que **deja de cumplirse la Ley de Darcy**  cuando $R_e$ alcanza un valor que vara entre 1 y 10. En este caso el caudal deja de ser linealmente proporcional al gradiente.

Si $R_{e}<1$, se cumple Darcy

Si $1>R_{e}<10$, no se cumple Darcy
