##  1. En la siguiente figura se representan la velocidades de sedimentación de diferentes partículas de arena que ingresan a un desarenador de 1.5 metros de altura y una longitud de 1.95 metros. Con base en esta información realice un gráfico en el que compare el número de Reynolds y el diámetro de cada partícula. Haga su respectivo análisis. Considere una temperatura del agua de 20°C y una densidad de partícula de 2500 $kg*m^{-3}$

```python
Vh = 390 #m/d
Vs = np.linspace(300,360,5) #m/d
```

```python
H = np.linspace(0,1.5,100)
```

```python
for i in range(len(Vs)):
    L = Vh*H/Vs[i]
    plt.plot(np.sort(L,axis=0)[::-1],H, label=np.round(Vs[i],2))
    plt.xlabel("Longitud del desarenador, m")
    plt.ylabel("Altura del desarenador, m")
    plt.legend(title="Velocidad de sedimentación, m/d")
```
