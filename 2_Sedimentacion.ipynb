# *SEDIMENTACIÓN*
<p style="font-size:40px ">  La sedimentación es un concepto que aplica dos operaciones unitarias, la clarificación y el espesamiento de sólidos</p> <br>

<p>La clarificación separa el proceso segun el tipo de partícula en dos clases.  La clarificación tipo-1, la cual corresponde a la remoción de partículas no floculentas. Clarificación tipo-2, que corresponde a la remoción de partículas floculentas.</p>

## **Clarificación tipo 1**

<p>Corresponde a la sedimentación de partículas discretas como las arenas o gravas </p>

```python
Angulos = np.linspace(0,360,1000)
AngIzq = np.linspace(90,92,1000)
AngDer = np.linspace(-90,-92,1000)
Radio = 1
x = np.cos(Angulos)*Radio
y = np.sin(Angulos)*Radio
xd = np.cos(AngDer)*1.5
yd = np.sin(AngDer)*1.5
xi = np.cos(AngIzq)*(-1.5)
yi = np.sin(AngIzq)*(-1.5)
plt.plot(x,y)
plt.plot(xd,yd, color="#00ff00")
plt.plot(xi,yi, color="#00ff00")
plt.ylim(top=5)
plt.ylim(bottom=-5)
plt.xlim(right=6)
plt.xlim(left= -5)
plt.arrow(0,4,0,-2, color="r", width=0.1)
plt.arrow(0,-4,0,2, color="b", width=0.1)
plt.arrow(1.5,0,0.5,0, color="g", width=0.1)
plt.annotate("Fuerza de gravedad", xy=(0.1,4))
plt.annotate("Fuerza de empuje", xy=(-3,-4))
plt.annotate("Fuerza de fricción \nActúa sobre el eje Y", xy=(2.5,0))
plt.show()
```

### Fuerzas que afectan la sedimentación de partículas tipo 1

$F_{Total} = F_{gravedad}- F_{empuje} - F_{fricción}$ <br>

$F_{gravedad} = {m_{partícula} \cdot g} = {\rho_{p} \cdot V_{p} \cdot g} $ <br>

$F_{empuje} = m_{f} \cdot g = V_{f} \cdot \rho_{f} \cdot g$<br>

$V_{f} = V_{p}$<br>

$F_{total}= 0$<br>

$F_{fricción}= {\frac {C_{D} \cdot A_{p} \cdot \rho_{f} \cdot \vec V^{2}}{2}}$<br>

$0 = {\rho_{p} \cdot V_{p} \cdot g}  - V_{f} \cdot \rho_{f} \cdot g  - {\frac {C_{D} \cdot A_{p} \cdot \rho_{f} \cdot \vec V^{2}}{2}}$ <br>

$0 = {\rho_{p} \cdot V_{p}}  - V_{p} \cdot \rho_{f} - {\frac {C_{D} \cdot A_{p} \cdot \rho_{f} \cdot \vec V^{2}}{2g}}$ <br>

$0 = V_{p} \cdot (\rho_{p} - \rho_{f}) - {\frac {C_{D} \cdot A_{p} \cdot \rho_{f} \vec V^{2}}{2g}}$ <br>

$0= {\frac {2g \cdot V_{p} \cdot (\rho_{p} - \rho_{f}) - C_{D} \cdot A_{p} \cdot \rho_{f} \cdot \vec V^{2}}{2g}}$ <br>

$0= 2g \cdot V_{p} \cdot (\rho_{p} - \rho_{f}) - C_{D} \cdot A_{p} \cdot \rho_{f} \cdot \vec V^{2}$ <br>

$C_{D} \cdot A_{p} \cdot \rho_{f} \vec V^{2}= 2g \cdot V_{p}\cdot (\rho_{p} - \rho_{f})$ <br>

$ \vec V^{2} = {\frac {2g V_{p}(\rho_{p} - \rho_{f})} {C_{D}  A_{p} \rho_{f}}}$ <br>

$V_{p} = {\frac {4 \pi r^{3}}{3}}$ <br>

$A_{p} = 4 \pi r^{2}$ <br>

$ \vec V^{2} = {\frac {8 \pi r^{3} g (\rho_{p} - \rho_{f})} {3 C_{D}  A_{p} \rho_{f}}}$ <br>

$ \vec V^{2} = {\frac {2g r (\rho_{p} - \rho_{f})} {3 C_{D}  \rho_{f}}}$ <br>

$ \vec V = \sqrt {{\frac {2g r (\rho_{p} - \rho_{f})} {3 C_{D}  \rho_{f}}} }$ <br>

$ \vec V = \sqrt {{\frac {2g r (\rho_{p} - \rho_{f})} {3 C_{D}  \rho_{f} \phi}} }$ <br> 

$\vec V = \sqrt { \frac {gd( \rho_{p} - \rho_{f} )} {3 C_{D} \rho_{f} \phi} }  $ <br> 
$x \times \cdot *$

### **Coeficiente de fricción**
1. Para Reynolds < 1:

$C_{D} = \frac {24} {N_{Re}}$

2. Para Reynolds > 10000:

$C_{D} = 0.34$

3. Para Retnold entre 1 y 10000:

$C_{D} = {\frac {24} {N_{Re}}} + {\frac {3} {\sqrt{N_{Re}}}} + 0.34$

4. Número de Reynolds:

$N_{Re} = {\frac {\rho_{f} \vec V d} {\mu}} = {\frac {\vec V d} {\upsilon}}$ <br>
$N_{Re} = {\frac {2 \rho_{f} \vec V r} {\mu}} = {\frac {2 \vec V r} {\upsilon}}$ <br>
Donde;<br>
$\mu$: viscosidad dinámica <br>
$\upsilon$: viscosidad cinemática 

### **Para Re < 1**
$ \vec V = \sqrt {{\frac {2g r (\rho_{p} - \rho_{f})} {3 C_{D}  \rho_{f} \phi}} }$ <br>

$ \vec V = \sqrt {{\frac {2g r N_{Re} (\rho_{p} - \rho_{f})} {3 * 24  \rho_{f} \phi}} }$ <br>

$ \vec V = \sqrt {{\frac {g r N_{Re} (\rho_{p} - \rho_{f})} {36  \rho_{f} \phi}} }$ <br>

$ \vec V = \sqrt {{\frac {g r N_{Re} (\rho_{p} - \rho_{f})} {36  \rho_{f} \phi}} }$ <br>

$ \vec V = \sqrt {{\frac {g r^{2} \vec V \rho_{f} (\rho_{p} - \rho_{f} )} {18  \mu \rho_{f} \phi}} }$ <br>

$ \vec V^{2} = {\frac {g r^{2} \vec V \rho_{f} (\rho_{p} - \rho_{f} )} {18  \mu \rho_{f} \phi}} $ <br>

$ \vec V  = {\frac {g r^{2}  \rho_{f} (\rho_{p} - \rho_{f})} {18  \mu \rho_{f} \phi}} $ <br>

$ \vec V  = {\frac {g r^{2}  (\rho_{p} - \rho_{f})} {18  \mu \phi}} $ <br>

$ \vec V  = {\frac {g r^{2}  (\rho_{p} - \rho_{f})} {18  \mu \phi}} $ <br>

### **Para Re > 10000**

$ \vec V = \sqrt {{\frac {2g r (\rho_{p} - \rho_{f})} {3 C_{D}  \rho_{f} \phi}} }$ <br>

$ \vec V = \sqrt {{\frac {g d (\rho_{p} - \rho_{f})} {3 * 0.34  \rho_{f} \phi}} }$ <br>

$ \vec V = \sqrt {{\frac {g d (\rho_{p} - \rho_{f})} {1.02  \rho_{f} \phi}} }$ <br>

$ \vec V = \sqrt {{\frac {g d (\rho_{p} - \rho_{f})} {1.02 \rho_{f} \phi}} }$ <br>

```python
PropiedadesAgua = {"Temperatura":[0,5,10,15,20,25,30,40,50,60,70,80,90,100], 
                   "Densidad":[999.8,1000,999.7,999.1,998.2,997,995.7,992.2,
                               988,983.2,977.8,971.8,965.3,958.4], 
                   "ViscosidadDim":[1.781e-3, 1.518e-3, 1.307e-3, 1.139e-3, 
                                    1.002e-3, 0.890e-3, 0.798e-3, 0.653e-3, 
                                    0.547e-3, 0.466e-3, 0.404e-3, 0.354e-3, 
                                    0.315e-3, 0.282e-3], 
                   "ViscosidadCin":[1.785e-6, 1.519e-6, 1.306e-6, 1.140e-6, 
                                    1.003e-6, 0.893e-6, 0.8e-6, 0.658e-6, 
                                    0.553e-6, 0.474e-6, 0.413e-6, 0.364e-6, 
                                    0.326e-6, 0.294e-6]}
PropiedadesAgua = pd.DataFrame(PropiedadesAgua,)
PropiedadesAgua = PropiedadesAgua.set_index(PropiedadesAgua["Temperatura"])
PropiedadesAgua

#Temperatura en grados C
#Densidad en kg/m3
#Viscosidad dinámica en Pa*s
#Viscosidad cinemática en m2/s
```

<h2 style="color:red;">1.1. Vamos a evaluar la sedimentación para diferentes tamaños de partículas a 20°C (para Re < 1)</h2>

```python
diametrosParticulas = np.linspace(1e-3, 0.01, 1000) #milimetros
radiosParticulas = diametrosParticulas/2
radiosParticulasM = radiosParticulas/1000 #metros
#radiosParticulasM
```

```python
FormaParticula = np.linspace(1.0,2.3,10)
FormaParticula = np.round(FormaParticula,2)
FormaParticula
```

```python
#Definir ecuación para NRe<1:

def VelSedRe1(r,dp,phi):
  df = PropiedadesAgua.loc[20, "Densidad"]
  mu = PropiedadesAgua.loc[20, "ViscosidadDim"]
  for i in np.arange(len(phi)):
    Vs = (9.81*(r**2)*(dp-df))/(18*mu*phi[i])
    plt.plot(r, Vs, label=phi[i])
    plt.xlabel("Radio de partícula, (m)")
    plt.ylabel("Velocidad de sedimentación (m/s)")
    plt.title("Velocidad de sedimentación por \ntamaño de partícula y forma a 20°V")
    plt.legend(title="Forma, $\phi$")
    #plt.show()
```

```python
VelSedRe1(radiosParticulasM, DensidadArenaCuarzo, FormaParticula)
```

<h3 style="color:red;">1.3. Vamos a evaluar el número de Reynolds para las velocidades de sedimentación estimadas y verificar si cumple con la condicion de un Re, para ello se asumirá una partícula ideal de forma esférica < 1</h3>

```python
#Definir ecuación para NRe<1:
# Evaluar para temperaturas del agua entre 0 y 100°C

def VelSedRe5(r,dp,phi):
  df = PropiedadesAgua["Densidad"]
  mu = PropiedadesAgua["ViscosidadDim"]
  temp =PropiedadesAgua["Temperatura"]
  for i in np.arange(len(r)):
    Vs = (9.81*(r[i]**2)*(dp-df))/(18*mu*1)
    Re = ((2*df*r[i]*Vs)/mu)
    plt.plot(Re, Vs, label=np.format_float_scientific(r[i],precision=2))
    plt.xlabel("Número de Reynolds")
    plt.ylabel("Velocidad de sedimentación (m/s)")
    plt.title("Velocidad de sedimentación \npor Número de Reynolds")
    plt.legend(title='Radio de parícula, $\mu m$')
    #plt.show()
```
